```python
def gene_reaction_list (path):
    genes = open (path, "r")
    gene_list = genes.read().split("\n")
    def gene_to_reaction(gene_list):
        reaction_list=[]
        for gene in gene_list:
            try:
                reaction_id=model.genes.get_by_id(gene).reactions
            except KeyError:
                continue
            else:
                for i in reaction_id:
                    reaction_list.append(i.id)
        return(reaction_list)
    reactions= list(set(gene_to_reaction(gene_list)))
    return(reactions)
    
```

Forward reactions to a .txt file

```python
import re

def reactions_to_file(reaction_list):
    my_variable=reaction_list
    variable_name = ""
    for name, value in globals().items():
        if value is my_variable:
            variable_name = name
    pattern="_list_genes"
    if re.search(pattern, variable_name):
        with open(variable_name.replace("_list_genes", "s")+"_operon.txt", 'w') as f:
            for i in reaction_list:
                f.write(i)
                f.write('\n')
    else:
        with open(variable_name.replace("_list", "s")+"_operon.txt", 'w') as f:
            for i in reaction_list:
                f.write(i)
                f.write('\n')
```

```python
import pandas as pd
def reaction_class(reactions,active_reactions_FBA, active_reactions_FVA05, active_reactions_FVA08,essential_reactions):
    data = {'reactions': reactions, 
    'FBA': ['active' if x else 'innactive' for x in [r in active_reactions_FBA for r in reactions]], 
    'FVA05': ['active' if x else 'innactive' for x in [r in active_reactions_FVA05 for r in reactions]],
    'FVA08': ['active' if x else 'innactive' for x in [r in active_reactions_FVA08 for r in reactions]], 
    'essentiality': ['essential' if x else 'non-essential' for x in [r in essential_reactions for r in reactions]]}
    df = pd.DataFrame(data)
    my_variable=reactions
    variable_name = ""
    for name, value in globals().items():
        if value is my_variable:
            variable_name = name
    df.to_csv(variable_name.replace("_list", "s")+"_summary.txt", sep='\t', index=False)
    return df
```

```python
def summary (reaction_list_active_solution_08,model_KO,model_KO_minimum_08,model_KO_maximum_08,
reaction_list_active_solution_05,model_KO_minimum_05,model_KO_maximum_05):
    wt_ko=pd.DataFrame({'reaction':list(model_KO.keys()),
    'bounds FBA':(solutionFBA.fluxes.loc[list(model_KO.keys())]).tolist(),
    'solutionFBA_KO':list(model_KO.values())
    })
    FVA08=pd.DataFrame({'reaction':reaction_list_active_solution_08.index,
    'min_bound_08':reaction_list_active_solution_08["minimum"].tolist(),
    "max_bound_08":reaction_list_active_solution_08["maximum"].tolist(),
    'solution_min_08':list(model_KO_minimum_08.values()),
    'solution_max_08':list(model_KO_maximum_08.values())})
    FVA05=pd.DataFrame({'reaction':reaction_list_active_solution_05.index,
    'min_bound_05':reaction_list_active_solution_05["minimum"].tolist(),
    "max_bound_05":reaction_list_active_solution_05["maximum"].tolist(),
    'solution_min_05':list(model_KO_minimum_05.values()),
    'solution_max_05':list(model_KO_maximum_05.values())
    })
    df1 = pd.merge(wt_ko, FVA05, on="reaction",how="outer")
    df2 = pd.merge(df1, FVA08, on="reaction",how="outer")
    df2['deviation_FBA_KO']=(abs(df2['solutionFBA_KO']-model.summary()._objective_value))
    df2['deviation_min_05']=(abs(df2['solution_min_05']-model.summary()._objective_value))
    df2['deviation_max_05']=(abs(df2['solution_max_05']-model.summary()._objective_value))
    df2['deviation_min_08']=(abs(df2['solution_min_08']-model.summary()._objective_value))
    df2['deviation_max_08']=(abs(df2['solution_max_08']-model.summary()._objective_value))
    df2.iloc[:,1:]=df2.iloc[:,1:].astype(float)
    return(df2)
```

```python
cobra_model=model.copy()

def double_gene_KO(model,non_essential_genes):
    double_KO_gene=double_gene_deletion(model,non_essential_genes)
    def get_reactions(gene_ids):
        reactions = set()
        for gene in gene_ids:
            gene_obj = model.genes.get_by_id(gene)
            reactions |= set(gene_obj.reactions)
        return list(reactions)
    
    double_KO_gene['reactions'] = double_KO_gene['ids'].apply(get_reactions)
    def get_reactions_ids(reaction_list):
        reactions = set()
        reactions_id=[]
        for n in range(len(reaction_list)):
            id=reaction_list[n].id
            reactions_id.append(id)
        return list(reactions_id)

    double_KO_gene['reactions_id'] = double_KO_gene['reactions'].apply(get_reactions_ids)
    double_KO_gene['reactions_id_sorted'] = double_KO_gene['reactions_id'].apply(lambda x: sorted(x))
    double_KO_reactions = double_KO_gene.drop_duplicates(subset='reactions_id_sorted', keep='first')[["growth","reactions_id_sorted"]]
    double_KO_reactions['growth'] = double_KO_reactions['growth'].round(6)
    double_KO_reactions = double_KO_reactions[double_KO_reactions['growth'] != 0.876997]

    def find_common_reaction(x):
        reactions = set(x.iloc[0])
        for r in x:
            reactions = reactions.intersection(set(r))
        return reactions

    common_reactions = double_KO_reactions.groupby('growth')['reactions_id_sorted'].apply(find_common_reaction)
    final_table = pd.DataFrame({'growth': common_reactions.index, 'common_reaction': common_reactions.values})

    return (final_table)
```

```python
import numpy as np
def deviation_reactions (reaction1,solutionFBA,reaction2=None):
    model_double_KO=model.copy()
    model_double_KO.reactions.get_by_id(reaction1).bounds=(0,0)
    if reaction2 is not None and reaction2 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction2).bounds = (0, 0)
    solutionKO=model_double_KO.optimize()
    names=[]
    distances= []
    fluxes=[]
    new_fluxes=[]
    percentaje=[]
    for flux in range(0,len(solutionKO.fluxes.index)):
        name=solutionKO.fluxes.index[flux]
        distance=solutionKO[flux] - solutionFBA[flux]
        if abs(distance) > abs(solutionFBA[flux])*0.1 and abs(distance) > 0.1:
            names.append(name)
            distances.append(round(distance,6))
            fluxes.append(round(solutionFBA[flux],6))
            new_fluxes.append(round(solutionKO[flux],6))
            if solutionFBA[flux] == 0:
                percent= np.nan
            else:
                percent=solutionKO[flux]*100/solutionFBA[flux]
            percentaje.append(round(percent,6))
    wt_vs_KO={"Reaction":names,"Deviation":distances,"flux":fluxes,"new_flux":new_fluxes,"percentaje":percentaje}
    df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation","percentaje","flux","new_flux"])
    df = df[df['Reaction'] != reaction1]
    df = df[df['Reaction'] != reaction2]
    df=df.set_index("Reaction")
    df = df.sort_values(
    by=['percentaje', 'Deviation'],
    ascending=[False, False],
    key=lambda col:col.abs(),
    na_position='first')
    return df
```

```python
import itertools
import numpy as np
```

```python
def deviation_reactions_pairwise (list_active,solutionFBA):
    combination= []
    for c in itertools.combinations(list_active, 2):
        combination.append(c)
    combinations_array = np.array(combination)
    lista=[]
    for i in range(1,len(combinations_array)):
        model_double_KO=model.copy()
        model_double_KO.reactions.get_by_id(combinations_array[i][0]).bounds=(0,0)
        model_double_KO.reactions.get_by_id(combinations_array[i][1]).bounds=(0,0)
        solutionKO=model_double_KO.optimize()
        names=[]
        distances= []
        for flux in range(0,len(solutionKO.fluxes.index)):
            name=solutionKO.fluxes.index[flux]
            distance=abs(solutionKO[flux] - solutionFBA[flux])
            if distance > abs(solutionFBA[flux])*0.1 and distance > 0.1:
                names.append(name)
                distances.append(distance)
        wt_vs_KO={"Reaction":names,"Deviation":distances}
        df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation"])
        df=df.set_index("Reaction").sort_values(by="Deviation")
        pair=[combinations_array[i][0],combinations_array[i][1]]
        lista.append(pair)
        lista.append(df)
    return lista
```
