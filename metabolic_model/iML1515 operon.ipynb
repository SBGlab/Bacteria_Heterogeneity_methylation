```python
def double_gene_KO(model,non_essential_genes):
    double_KO_gene=double_gene_deletion(model,non_essential_genes)
    def get_reactions(gene_ids):
        reactions = set()
        for gene in gene_ids:
            gene_obj = model.genes.get_by_id(gene)
            reactions |= set(gene_obj.reactions)
        return list(reactions)
    
    double_KO_gene['reactions'] = double_KO_gene['ids'].apply(get_reactions)
    def get_reactions_ids(reaction_list):
        reactions = set()
        reactions_id=[]
        for n in range(len(reaction_list)):
            id=reaction_list[n].id
            reactions_id.append(id)
        return list(reactions_id)

    double_KO_gene['reactions_id'] = double_KO_gene['reactions'].apply(get_reactions_ids)
    double_KO_gene['reactions_id_sorted'] = double_KO_gene['reactions_id'].apply(lambda x: sorted(x))
    double_KO_reactions = double_KO_gene.drop_duplicates(subset='reactions_id_sorted', keep='first')[["growth","reactions_id_sorted"]]
    double_KO_reactions['growth'] = double_KO_reactions['growth'].round(6)
    double_KO_reactions = double_KO_reactions[double_KO_reactions['growth'] != 0.876997]

    def find_common_reaction(x):
        reactions = set(x.iloc[0])
        for r in x:
            reactions = reactions.intersection(set(r))
        return reactions

    common_reactions = double_KO_reactions.groupby('growth')['reactions_id_sorted'].apply(find_common_reaction)
    final_table = pd.DataFrame({'growth': common_reactions.index, 'common_reaction': common_reactions.values})

    return (final_table)
```

```python
import numpy as np
def deviation_reactions (reaction1,solutionFBA,reaction2=None):
    model_double_KO=model.copy()
    model_double_KO.reactions.get_by_id(reaction1).bounds=(0,0)
    if reaction2 is not None and reaction2 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction2).bounds = (0, 0)
    solutionKO=model_double_KO.optimize()
    names=[]
    distances= []
    fluxes=[]
    new_fluxes=[]
    percentaje=[]
    for flux in range(0,len(solutionKO.fluxes.index)):
        name=solutionKO.fluxes.index[flux]
        distance=solutionKO[flux] - solutionFBA[flux]
        if abs(distance) > abs(solutionFBA[flux])*0.1 and abs(distance) > 0.1:
            names.append(name)
            distances.append(round(distance,6))
            fluxes.append(round(solutionFBA[flux],6))
            new_fluxes.append(round(solutionKO[flux],6))
            if solutionFBA[flux] == 0:
                percent= np.nan
            else:
                percent=solutionKO[flux]*100/solutionFBA[flux]
            percentaje.append(round(percent,6))
    wt_vs_KO={"Reaction":names,"Deviation":distances,"flux":fluxes,"new_flux":new_fluxes,"percentaje":percentaje}
    df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation","percentaje","flux","new_flux"])
    df = df[df['Reaction'] != reaction1]
    df = df[df['Reaction'] != reaction2]
    df=df.set_index("Reaction")
    df = df.sort_values(
    by=['percentaje', 'Deviation'],
    ascending=[False, False],
    key=lambda col:col.abs(),
    na_position='first')
    return df
```

```python
import itertools
import numpy as np
```

```python
def deviation_reactions_pairwise (list_active,solutionFBA):
    combination= []
    for c in itertools.combinations(list_active, 2):
        combination.append(c)
    combinations_array = np.array(combination)
    lista=[]
    for i in range(1,len(combinations_array)):
        model_double_KO=model.copy()
        model_double_KO.reactions.get_by_id(combinations_array[i][0]).bounds=(0,0)
        model_double_KO.reactions.get_by_id(combinations_array[i][1]).bounds=(0,0)
        solutionKO=model_double_KO.optimize()
        names=[]
        distances= []
        for flux in range(0,len(solutionKO.fluxes.index)):
            name=solutionKO.fluxes.index[flux]
            distance=abs(solutionKO[flux] - solutionFBA[flux])
            if distance > abs(solutionFBA[flux])*0.1 and distance > 0.1:
                names.append(name)
                distances.append(distance)
        wt_vs_KO={"Reaction":names,"Deviation":distances}
        df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation"])
        df=df.set_index("Reaction").sort_values(by="Deviation")
        pair=[combinations_array[i][0],combinations_array[i][1]]
        lista.append(pair)
        lista.append(df)
    return lista
```
