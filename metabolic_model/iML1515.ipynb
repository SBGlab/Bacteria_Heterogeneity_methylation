Extracting reactions associated to differentially methylated genes

```python
def gene_to_reaction (gene_list):
    reaction_list=[]
    for gene in gene_list:
        try:
            reaction_id=model.genes.get_by_id(gene).reactions
        except KeyError:
            continue
        else:
            for i in reaction_id:
                reaction_list.append(i.id)
    return(reaction_list)
```

```python
def model_KO_solution(reaction_list_active):
    model_KO={}
    for reaction in reaction_list_active:
        model_copy=model.copy()
        model_copy.reactions.get_by_id(reaction).bounds=(0,0)
        model_KO[reaction]=model_copy.optimize().objective_value
    return(model_KO)

```

Flux deviations

```python
def deviation_reactions (gene1,gene2,solutionFBA):
    model_double_KO=model.copy()
    model_double_KO.reactions.get_by_id(gene1).bounds=(0,0)
    model_double_KO.reactions.get_by_id(gene2).bounds=(0,0)
    solutionKO=model_double_KO.optimize()
    names=[]
    distances= []
    for flux in range(0,len(solutionKO.fluxes.index)):
        name=solutionKO.fluxes.index[flux]
        distance=solutionKO[flux] - solutionFBA[flux]
        if distance > abs(0.01):
            names.append(name)
            distances.append(distance)
    wt_vs_KO={"Reaction":names,"Deviation":distances}
    df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation"])
    df=df.set_index("Reaction").sort_values(by="Deviation")
    return df
```
