```python
#Forward single reactions to file and variable
import re

def gene_read(path):
    gene_list=[]
    with open (path,"r") as file:
        for line in file:
            line_elements = line.strip().split(',')
            gene_list.extend(line_elements)
    return(gene_list)


def gene_reaction_list (path):
    gene_list=[]
    with open (path,"r") as file:
        for line in file:
            line_elements = line.strip().split(',')
            gene_list.extend(line_elements)
    def gene_to_reaction(gene_list):
        associated_reactions = []
        for gene_id in gene_list:
            gene_obj = model.genes.get_by_id(gene_id)
            reactions = [reaction.id for reaction in gene_obj.reactions]
            associated_reactions.extend(reactions)
        return(associated_reactions)
    reactions= list(set(gene_to_reaction(gene_list)))
    start_index = path.index("_genelist")
    end_index = path.index("operon_")+len("operon_")
    substring = path[end_index:start_index]
    with open("reactions_"+ substring +"_operon.txt", 'w') as f:
        for i in reactions:
            f.write(i)
            f.write('\n')
    return(reactions)
```

```python
def single_gene_KO(model,non_essential_genes):
    single_KO_gene=single_gene_deletion(model,non_essential_genes)
    def get_associated_reactions(gene_ids):
        associated_reactions = []
        for gene_set in gene_ids:
            gene_list = list(gene_set)
            gene_reactions = []
            for gene in gene_list:
                gene_obj = model_M9.genes.get_by_id(gene)
                gene_reactions.extend([reaction.id for reaction in gene_obj.reactions])
            associated_reactions.append(','.join(gene_reactions))
        return associated_reactions


    single_KO_gene['associated_reactions'] = get_associated_reactions(single_KO_gene['ids'])
    single_KO_gene['growth'] = single_KO_gene['growth'].round(6)
    def concat_gene_ids(ids):
        return list(set.union(*ids))
    summarized_single_KO_gene = single_KO_gene.groupby(['growth', 'associated_reactions'], as_index=False).agg({'ids': concat_gene_ids})
    summarized_single_KO_gene = summarized_single_KO_gene.rename(columns={'ids': 'genes'})
    summarized_single_KO_gene['associated_reactions'] = summarized_single_KO_gene['associated_reactions'].apply(lambda x: [r for r in x.split(',') if r.strip() in active_reactionsFBA])
    summarized_single_KO_gene = summarized_single_KO_gene[summarized_single_KO_gene['associated_reactions'].str.len() > 0]
    return(summarized_single_KO_gene)
    
```

Double GENE KO, associated reactions

```python
#Doing the KO of two genes, depending on the GPR, the associated reaction/s will be set to bounds=0. Retrieve these associated reactions
#when the resulting growth rate of two or more  double gene KOs is identical compare the KO reactions associated and find the common one/s
#in all cases and retrieve only once the growth rate of the associated reaction/s KO. Do not retrieve the KOs that don't influience the 
#growth rate
def double_gene_KO(model,non_essential_genes):
    double_KO_gene=double_gene_deletion(model,non_essential_genes)
    def get_reactions(gene_ids):
        reactions = set()
        for gene in gene_ids:
            gene_obj = model.genes.get_by_id(gene)
            reactions |= set(gene_obj.reactions)
        return list(reactions)

    double_KO_gene['reactions'] = double_KO_gene['ids'].apply(get_reactions)
    def get_reactions_ids(reaction_list):
        reactions_id=[]
        for n in range(len(reaction_list)):
            id=reaction_list[n].id
            if id in active_reactionsFBA:
                reactions_id.append(id)
        return list(reactions_id)
    double_KO_gene['growth'] = double_KO_gene['growth'].round(4)
    double_KO_gene['reactions_id'] = double_KO_gene['reactions'].apply(get_reactions_ids)
    double_KO_gene['reactions_id_sorted'] = double_KO_gene['reactions_id'].apply(lambda x: sorted(x))
    double_KO_gene['reactions_id_sorted_tuple'] = double_KO_gene['reactions_id_sorted'].apply(tuple)
    double_KO_reactions = double_KO_gene.groupby('reactions_id_sorted_tuple').agg({'growth': 'first', 'ids': lambda x: list(x.values)})
    double_KO_reactions = double_KO_reactions.reset_index()  # Reset the index to make the grouped column a regular column
    double_KO_reactions = double_KO_reactions[['reactions_id_sorted_tuple', 'growth', 'ids']]  # Select the desired columns
    double_KO_reactions = double_KO_reactions[double_KO_reactions['growth'] != 0.8770]
    def find_common_reaction(x):
            reactions = set(x.iloc[0])
            for r in x:
                reactions = reactions.intersection(set(r))
            return reactions

    common_reactions = double_KO_reactions.groupby('growth')['reactions_id_sorted_tuple'].apply(find_common_reaction)
    common_reactions = common_reactions.reset_index()
    grouped_ids = double_KO_reactions.groupby('growth')['ids'].apply(list).reset_index()
    result = common_reactions.merge(grouped_ids, on='growth')
    result['ids'] = result['ids'].apply(lambda x: x[0])
    return(result)
```

```python
def gene_upperbound(gene_list,solution):
    results = pd.DataFrame()

    for gene in gene_list:
        reaction_list = []
        reaction_id = model_M9.genes.get_by_id(gene).reactions

        for i in reaction_id:
            reaction_list.append(i.id)

        reactions = list(set(reaction_list))
        active_reactions = []

        for i in reactions:
            if i in solution.index:
                active_reactions.append(solution.loc[i])

        model_copy = model_M9.copy()

        if len(active_reactions) == 0:
            continue
        else:
            for i in active_reactions:
                max_bound = max(abs(i["maximum"]), abs(i["minimum"]))
                model_copy.reactions.get_by_id(i.name).bounds = (max_bound, max_bound)

            model_copy.optimize()
            objective_value = model_copy.objective.value

            for i in active_reactions:
                data = {"gene": gene, "reaction": i.name, "bound": max(abs(i["maximum"]), abs(i["minimum"])).round(4), "solution": objective_value}
                df = pd.DataFrame(data, index=[0])
                results = pd.concat([results, df])
    results_collapsed = results.groupby("gene").agg(lambda x: x.tolist() if x.nunique() > 1 else x.iloc[0]).reset_index()
    return(results_collapsed)
```

```python
import numpy as np
def deviation_reactions (reaction1,solutionFBA,reaction2=None, reaction3=None,reaction4=None,reaction5=None,reaction6=None):
    model_double_KO=model_M9.copy()
    model_double_KO.reactions.get_by_id(reaction1).bounds=(0,0)
    if reaction2 is not None and reaction2 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction2).bounds = (0, 0)
    if reaction3 is not None and reaction3 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction3).bounds = (0, 0)
    if reaction4 is not None and reaction4 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction4).bounds = (0, 0)
    if reaction5 is not None and reaction5 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction5).bounds = (0, 0)
    if reaction6 is not None and reaction6 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction6).bounds = (0, 0)
    solutionKO=model_double_KO.optimize()
    names=[]
    distances= []
    fluxes=[]
    new_fluxes=[]
    percentaje=[]
    for flux in range(0,len(solutionKO.fluxes.index)):
        name=solutionKO.fluxes.index[flux]
        distance=solutionKO[flux] - solutionFBA[flux]
        if abs(distance) > abs(solutionFBA[flux])*0.1 and abs(distance) > 0.1:
            names.append(name)
            distances.append(round(distance,6))
            fluxes.append(round(solutionFBA[flux],6))
            new_fluxes.append(round(solutionKO[flux],6))
            if solutionFBA[flux] == 0:
                percent= np.nan
            else:
                percent=solutionKO[flux]*100/solutionFBA[flux]
            percentaje.append(round(percent,6))
    wt_vs_KO={"Reaction":names,"Deviation":distances,"flux":fluxes,"new_flux":new_fluxes,"percentaje":percentaje}
    df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation","percentaje","flux","new_flux"])
    df = df[df['Reaction'] != reaction1]
    df = df[df['Reaction'] != reaction2]
    df = df[df['Reaction'] != reaction3]
    df = df[df['Reaction'] != reaction4]
    df = df[df['Reaction'] != reaction5]
    df = df[df['Reaction'] != reaction6]   
    df=df.set_index("Reaction")
    df = df.sort_values(
    by=['Deviation', 'percentaje'],
    ascending=[False, False],
    key=lambda col:col.abs(),
    na_position='first')
    return df
```

```python
import itertools
import numpy as np
```

```python
def deviation_reactions_pairwise (list_active,solutionFBA):
    combination= []
    for c in itertools.combinations(list_active, 2):
        combination.append(c)
    combinations_array = np.array(combination)
    lista=[]
    for i in range(1,len(combinations_array)):
        model_double_KO=model_M9.copy()
        model_double_KO.reactions.get_by_id(combinations_array[i][0]).bounds=(0,0)
        model_double_KO.reactions.get_by_id(combinations_array[i][1]).bounds=(0,0)
        solutionKO=model_double_KO.optimize()
        names=[]
        distances= []
        for flux in range(0,len(solutionKO.fluxes.index)):
            name=solutionKO.fluxes.index[flux]
            distance=abs(solutionKO[flux] - solutionFBA[flux])
            if distance > abs(solutionFBA[flux])*0.1 and distance > 0.1:
                names.append(name)
                distances.append(distance)
        wt_vs_KO={"Reaction":names,"Deviation":distances}
        df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation"])
        df=df.set_index("Reaction").sort_values(by="Deviation")
        pair=[combinations_array[i][0],combinations_array[i][1]]
        lista.append(pair)
        lista.append(df)
    return lista
```
