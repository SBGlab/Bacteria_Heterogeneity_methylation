```python
#Forward single reactions to file and variable
import re

def gene_read(path):
    gene_list=[]
    with open (path,"r") as file:
        for line in file:
            line_elements = line.strip().split(',')
            gene_list.extend(line_elements)
    return(gene_list)


def gene_reaction_list (path):
    gene_list=[]
    with open (path,"r") as file:
        for line in file:
            line_elements = line.strip().split(',')
            gene_list.extend(line_elements)
    def gene_to_reaction(gene_list):
        associated_reactions = []
        for gene_id in gene_list:
            gene_obj = model.genes.get_by_id(gene_id)
            reactions = [reaction.id for reaction in gene_obj.reactions]
            associated_reactions.extend(reactions)
        return(associated_reactions)
    reactions= list(set(gene_to_reaction(gene_list)))
    start_index = path.index("_genelist")
    end_index = path.index("cobra_operon\\")+len("cobra_operon\\")
    substring = path[end_index:start_index]
    with open("reactions_"+ substring +"_operon.txt", 'w') as f:
        for i in reactions:
            f.write(i)
            f.write('\n')
    return(reactions)
```

```python
def active_solution(solution):
    active_reactions_list=[]
    for i in range(0,len(solution)):
        if abs(solution["minimum"].values[i]) >=0.0001 or abs(solution["maximum"].values[i]) >=0.0001:
            active_reactions_list.append(solution.index[i])
    return(active_reactions_list)
```

```python
import pandas as pd
def reaction_class(reactions,active_reactions_FBA, active_reactions_FVA05, active_reactions_FVA08,essential_reactions,output_name):
    data = {'reactions': reactions, 
    'FBA': ['active' if x else 'innactive' for x in [r in active_reactions_FBA for r in reactions]],
    'FVA08': ['active' if x else 'innactive' for x in [r in active_reactions_FVA08 for r in reactions]],     
    'FVA05': ['active' if x else 'innactive' for x in [r in active_reactions_FVA05 for r in reactions]],
    'essentiality': ['essential' if x else 'non-essential' for x in [r in essential_reactions for r in reactions]]}
    df = pd.DataFrame(data)
    df.to_csv(str(output_name) +".txt", sep='\t', index=False)
    return df
```

Double GENE KO, associated reactions

```python
#Doing the KO of two genes, depending on the GPR, the associated reaction/s will be set to bounds=0. Retrieve these associated reactions
#when the resulting growth rate of two or more  double gene KOs is identical compare the KO reactions associated and find the common one/s
#in all cases and retrieve only once the growth rate of the associated reaction/s KO. Do not retrieve the KOs that don't influience the 
#growth rate

cobra_model=model_LB.copy()

def double_gene_KO(model_LB,non_essential_genes,active_reactions):
    double_KO_gene=double_gene_deletion(model,non_essential_genes)
    def get_reactions(gene_ids):
        reactions = set()
        for gene in gene_ids:
            gene_obj = model_LB.genes.get_by_id(gene)
            reactions |= set(gene_obj.reactions)
        return list(reactions)
    
    double_KO_gene['reactions'] = double_KO_gene['ids'].apply(get_reactions)
    def get_reactions_ids(reaction_list):
        reactions = set()
        reactions_id=[]
        for n in range(len(reaction_list)):
            id=reaction_list[n].id
            if id in active_reactions:
                reactions_id.append(id)
        return list(reactions_id)

    double_KO_gene['reactions_id'] = double_KO_gene['reactions'].apply(get_reactions_ids)
    double_KO_gene['reactions_id_sorted'] = double_KO_gene['reactions_id'].apply(lambda x: sorted(x))
    double_KO_reactions = double_KO_gene.drop_duplicates(subset='reactions_id_sorted', keep='first')[["growth","reactions_id_sorted"]]
    double_KO_reactions['growth'] = double_KO_reactions['growth'].round(6)
    double_KO_reactions = double_KO_reactions[double_KO_reactions['growth'] != 0.876997]

    def find_common_reaction(x):
        reactions = set(x.iloc[0])
        for r in x:
            reactions = reactions.intersection(set(r))
        return reactions

    common_reactions = double_KO_reactions.groupby('growth')['reactions_id_sorted'].apply(find_common_reaction)
    final_table = pd.DataFrame({'growth': common_reactions.index, 'reaction/s': common_reactions.values})

    return (final_table)
```

```python
import numpy as np
def deviation_reactions (reaction1,solutionFBA,reaction2=None, reaction3=None,reaction4=None,reaction5=None,reaction6=None):
    model_double_KO=model_LB.copy()
    model_double_KO.reactions.get_by_id(reaction1).bounds=(0,0)
    if reaction2 is not None and reaction2 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction2).bounds = (0, 0)
    if reaction3 is not None and reaction3 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction3).bounds = (0, 0)
    if reaction4 is not None and reaction4 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction4).bounds = (0, 0)
    if reaction5 is not None and reaction5 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction5).bounds = (0, 0)
    if reaction6 is not None and reaction6 in model_double_KO.reactions:
        model_double_KO.reactions.get_by_id(reaction6).bounds = (0, 0)
    solutionKO=model_double_KO.optimize()
    names=[]
    distances= []
    fluxes=[]
    new_fluxes=[]
    percentaje=[]
    for flux in range(0,len(solutionKO.fluxes.index)):
        name=solutionKO.fluxes.index[flux]
        distance=solutionKO[flux] - solutionFBA[flux]
        if abs(distance) > abs(solutionFBA[flux])*0.1 and abs(distance) > 0.1:
            names.append(name)
            distances.append(round(distance,6))
            fluxes.append(round(solutionFBA[flux],6))
            new_fluxes.append(round(solutionKO[flux],6))
            if solutionFBA[flux] == 0:
                percent= np.nan
            else:
                percent=solutionKO[flux]*100/solutionFBA[flux]
            percentaje.append(round(percent,6))
    wt_vs_KO={"Reaction":names,"Deviation":distances,"flux":fluxes,"new_flux":new_fluxes,"percentaje":percentaje}
    df=pd.DataFrame(wt_vs_KO, columns=["Reaction","Deviation","percentaje","flux","new_flux"])
    df = df[df['Reaction'] != reaction1]
    df = df[df['Reaction'] != reaction2]
    df = df[df['Reaction'] != reaction3]
    df = df[df['Reaction'] != reaction4]
    df = df[df['Reaction'] != reaction5]
    df = df[df['Reaction'] != reaction6]   
    df=df.set_index("Reaction")
    df = df.sort_values(
    by=['Deviation', 'percentaje'],
    ascending=[False, False],
    key=lambda col:col.abs(),
    na_position='first')
    return df
```
